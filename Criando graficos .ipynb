```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
genero = ["Rock", "Hip Hop", "Pop", "Jazz"]
quantidade_msc = [570 , 465 , 845 , 916]

# Fundo Negro (ECA)
plt.style.use("dark_background")
# Desativar o plt.dark = plt.style.use("default")

cores = sns.color_palette("gnuplot2")

#Fonte "Minha cabea"
fig = plt.figure(figsize=(11,11))
plt.title("Ritmos onde os psicopatas mais escutam", fontsize=(20))
plt.pie(quantidade_msc, labels=genero, autopct='%1.1f%%', colors=(cores))
plt.show()
```

```python
jogadores = ["velho lindo" , "O que cria cavalo" , "sem tornoza" , "O que ficou famoso pq enterra"]
pontos = [21 , 29 , 17, 15]

plt.style.use("dark_background")

fig = plt.figure(figsize=(11,11))
plt.title("Pontos na varzia", fontsize=(20))
plt.pie(pontos, labels=jogadores, autopct='%1.1f%%', colors=(cores))
plt.show()
```
